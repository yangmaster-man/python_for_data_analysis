### 常见数据格式1
    ——不同格式读取方式不同（以什么样的后缀结尾，其中内容遵循怎样的格式规则）
#### JSON—JavaScript object notation—java对象表示法
    其信息结构特点使其能够很容易的转化为其他编程语言结构
##### 两种语言结构：
    json对象{}键值对-python字典；json数组[]值-python列表）
##### 特别语言规则：
     json的键：只可以用字符串，且字符串必须被双引号包围，键不可重复；
     json的值：只可以是字符串（双引号）、数字、布尔值（小写开头）、数组（支持嵌套）、对象（同上）、空值
##### JSON数据 解析转换 Python数据：
    pandas库的支持。pd.read_json() 文件读取、json解析、转成DataFrame全流程

### 常见数据格式2——CSV
#### CSV—comma separated values—逗号分隔值—数据分析师最爱
    第一行 通常为标头
    本身是一个非常规整的二维结构
    csv、json都是纯文本格式
    csv——excel or 记事本 直接写
    但是现实操作中，1g的很多，最好不用excel打开，卡到崩溃。
##### 优势：
    体积小 结构工整，容易让人理解
    可以直接转换成表格，也可以用excel表格直接打开、修改与导出
##### CSV数据 解析转换 Python数据：
    pandas库的支持。pd.read_csv() 文件读取、解析、转成DataFrame全流程
