**更新一列值**

```python
import pandas as pd
```

```python
name=pd.Series(["小城","小李","小张","小汪","小找","小轴"],index=["001","002","003","004","005","006"])
gender=pd.Series(["女","女","男","男","女","男"],
                index=["006","005","004","003","002","001"])
height=pd.Series([172.5,168.0,178.2,181.3,161.7],
                index=["001","002","003","004","005"])
students=pd.DataFrame({"姓名":name,"性别":gender,"身高":height})
students
```

**Series赋值**——如果被重新赋值的Series，原来有index，则新赋值的也要有——否则无法对其，产生值缺失NAN

```python
students.性别=pd.Series(["女","女","女","女","女","男"],
                      index=["006","005","004","003","002","001"])
```

```python
students
```

**列表赋值**——新列只可以用 students['成绩']，不能用 students.性别

```python
students.性别=["女","女","女","女","女","女"]
students.性别
```

**给DataFrame增加一列值**

```python
students['成绩']=[180,179,172,168,190,181]
students
```

**给DataFrame增加一行值**——只能用loc（iloc无用）——所以行的位置索引必须存在

```python
students.loc["007"] = ["小样","男","180","400"]
students
```

```python
students.loc["005"] = ["小明","男","180","400"]
students
```

**删除DataFrame的行或者列**——drop（）只返回值，不改变值；如果要改变值则需要重新赋值

```python
students.drop(["003"])
```

```python
students.drop(["001","005"])
```

```python
students=students.drop(["001"])
students
```

```python
students.drop("身高",axis=1)
```

students.drop("身高",axis=1)——DAtaFrame是二维数组：axis值为0时，为纵向，为1时，为横向——默认为零

```python
students.drop(["身高","成绩"],axis=1)
```
